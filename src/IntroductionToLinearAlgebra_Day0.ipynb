```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000)
```

Distance Travelled by robber [s=d/t] d=2.5t

```python
d_r = 2.5*t
```

Distabce travelled by Serrif d=3(t-5)

```python
d_s=3*(t-5)
```

```python
fig, ax = plt.subplots()
plt.title('A bank robber caught')
plt.xlabel('Time in minutes')
plt.ylabel('Distance in Km')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t, d_r, c='red')
ax.plot(t, d_s, c='yellow')

plt.axvline(x=30, color='purple', linestyle='--')
plt.axhline(y=75, color='purple', linestyle='--')
```
